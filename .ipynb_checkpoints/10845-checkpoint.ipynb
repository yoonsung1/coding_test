```python
import sys
input = sys.stdin.readline
```

```python
from collections import deque

N = int(input())
que = deque()
for i in range(N) :
    order = input().split()
    if len(order) == 2 :
        que.append(order[1])
    else :
        order = order[0]
        if order == 'pop' :
            if que :
                print(que.popleft())
            else :
                print(-1)
        elif order == 'size' :
            print(len(que))
        elif order == 'empty' :
            if que :
                print(0)
            else :
                print(1)
        elif order == 'front' :
            if que :
                print(que[0])
            else :
                print(-1)
        elif order == 'back' :
            if que :
                print(que[-1])
            else :
                print(-1)
```
