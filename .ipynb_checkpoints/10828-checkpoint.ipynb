```python
N = int(input())
stack = []
for i in range(N) :
    order = input().split()
    if len(order) == 2 :
        stack.append(order[1])
    elif order[0] == 'pop' :
        if not stack :
            print(-1)
        else :
            temp = stack.pop()
            print(temp)
    elif order[0] == 'size' :
        print(len(stack))
    elif order[0] == 'empty' :
        if not stack :
            print(1)
        else :
            print(0)
    elif order[0] == 'top' :
        if not stack :
            print(-1)
        else :
            print(stack[-1])
            
```
